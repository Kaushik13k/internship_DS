```python
import spacy
nlp = spacy.load("en_core_web_sm")
from spacy import displacy
```

```python
text = nlp(""" Apple is a Fruit. Vicky likes apple. Apple is also a Tech brand. Apple headoffice is in California.
           Apple invested 100 billion in india. Apple woring hours starts feom 9 IST """)
```

```python
for token in text:
    print(token.text, "\t\t", token.lemma_)
```

```python
#getting entities
for token in text.ents:
    print(token.text,"\t\t",token.label_)
```

```python
displacy.render(text, style='ent', jupyter=True) #entity Graph
```

```python
displacy.render(text, style='dep', jupyter=True) #dependency Graph
```
