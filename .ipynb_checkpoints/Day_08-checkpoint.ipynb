## pandas dataTypes:
            ### -Series       -1D
            ### -Dataframes   -2D
            ### -Panel        -3D

```python
#series
s1 = pd.Series([1,2,3,4,np.nan,5])   #np.nan---> create null value
print(type(s1))
print(s1) #Default the index starts from 0,1,2,3,4,...
print('//////////////////////////////////////////////////////////////////')
s2 = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
print(s2)
print('//////////////////////////////////////////////////////////////////')
s3 = pd.Series((1,2,3,4,5), index=('a','b','c','d','e'))
print(s3)
print('//////////////////////////////////////////////////////////////////')
# s4.name = "Dictionary Series"
# print(s4)
s4 = pd.Series({'a':1.11,'b':2,'c':3,'d':4,'e':5})
s4.index.name = "D.I"
print(s4)
print(f"the dimention is :{s4.ndim}")
```

```python
print(dir(pd.Series))
```

```python
#Dataframes
d1 = pd.DataFrame({'a':1, 'b':2,'c':3}) # see the error "If using all scalar values, you must pass an index"
```

```python
d1 = pd.DataFrame({'a':[1], 'b':[2],'c':[3]})
print(d1)
print(f"the dimention is --->{d1.ndim}")
print('//////////////////////////////////////////////////////////////////')
d2 = pd.DataFrame([1,2,3])
print(d2)
print('//////////////////////////////////////////////////////////////////')
d3 = pd.DataFrame([1,2,3], index = ['a', 'b', 'c'])
print(d3)
print(f"the size is -->{d2.size}")
print(f"the info is -->{d2.info()}")
print('//////////////////////////////////////////////////////////////////')
print("memory",d2.memory_usage())
```
